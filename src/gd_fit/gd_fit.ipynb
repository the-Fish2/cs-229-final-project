```python
%load_ext autoreload
%autoreload 2

import numpy as np
import sympy as sp
from utils.function_utils import gen_random_example, XS
from gd_fit import fit_best_function, fit_expr
import tensorflow as tf
from concurrent.futures import ThreadPoolExecutor, as_completed

print(tf.config.list_physical_devices('XPU'))
```

```python

# Generate a target function
RNG = np.random.default_rng(127)
f_expr = gen_random_example(RNG, complexity=2)
f_sampled = np.array([f_expr.subs(sp.symbols('x'), xi) for xi in XS], dtype=np.float64)
print('Sampled function: ', f_expr)

a0, a1, a2, b0, b1, b2, x = sp.symbols('a0, a1, a2, b0, b1, b2, x')
template = a1 * x + b1 + a2 * sp.sin(a0 * x + b0) + b2
# Fit best function
# best_expr, best_params, best_loss = fit_best_function(f_sampled, 'img2', search_complexity=1, hints=['sin'])
best_expr, best_params, best_loss = fit_expr(template, f_sampled, 'img2', hints=['sin'])
print("Best expression:", best_expr)
print("Best parameters:", best_params)
```

```python
# Generate a target function
RNG = np.random.default_rng(125)
f_expr = gen_random_example(RNG, complexity=1)
f_sampled = np.array([f_expr.subs(sp.symbols('x'), xi) for xi in XS], dtype=np.float64)
print('Sampled function: ', f_expr)

a0, a1, a2, b0, b1, b2, x = sp.symbols('a0, a1, a2, b0, b1, b2, x')
template = a1 * sp.cos(a0 * x + b0) + b1
# Fit best function
# best_expr, best_params, best_loss = fit_best_function(f_sampled, 'img2', search_complexity=1, hints=['sin'])
best_expr, best_params, best_loss = fit_expr(template, f_sampled, 'img1', hints=['cos'])
print("Best expression:", best_expr)
print("Best parameters:", best_params)
```
